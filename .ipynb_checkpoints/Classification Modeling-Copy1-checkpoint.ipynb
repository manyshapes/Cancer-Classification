```python
# Helper function for printing confusion matrices (see: https://gist.github.com/shaypal5/94c53d765083101efc0240d776a23823)

# prints confusion matrix as a heatmap which is nicer to visaulize

def print_confusion_matrix(confusion_matrix, class_names, figsize = (10,7), fontsize=18):
    """Prints a confusion matrix, as returned by sklearn.metrics.confusion_matrix, as a heatmap.
    
    Arguments
    ---------
    confusion_matrix: numpy.ndarray
        The numpy.ndarray object returned from a call to sklearn.metrics.confusion_matrix. 
        Similarly constructed ndarrays can also be used.
    class_names: list
        An ordered list of class names, in the order they index the given confusion matrix.
    figsize: tuple
        A 2-long tuple, the first value determining the horizontal size of the ouputted figure,
        the second determining the vertical size. Defaults to (10,7).
    fontsize: int
        Font size for axes labels. Defaults to 14.
        
    Returns
    -------
    matplotlib.figure.Figure
        The resulting confusion matrix figure
    """
    df_cm = pd.DataFrame(confusion_matrix, index=class_names, columns=class_names, )
    fig = plt.figure(figsize=figsize)
    try:
        heatmap = sns.heatmap(df_cm, annot=True, fmt="d")
    except ValueError:
        raise ValueError("Confusion matrix values must be integers.")
    heatmap.yaxis.set_ticklabels(heatmap.yaxis.get_ticklabels(), rotation=0, ha='right', fontsize=fontsize)
    heatmap.xaxis.set_ticklabels(heatmap.xaxis.get_ticklabels(), rotation=45, ha='right', fontsize=fontsize)
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    return fig
```

## H2O AI Models
Written in Java [H2o.ai](http://docs.h2o.ai/h2o/latest-stable/h2o-docs/index.html) models work well on big data by reading data parallel in clusters & storing compressed. 

```python
import h2o
from h2o.estimators import H2ORandomForestEstimator
```

```python
h2o.init()

# Import the cars dataset into H2O:
h2oset = h2o.import_file('75edits.csv')
```

```python
h2oset.summary
```

### Setting predictors & Response
Set the predictors and response.

Set the response as a factor.
- response needs to be set as a factor.

```python
# Setting x & y
predictors = ['N_Tumors','MO_Survival','DX_Year', 'AGE_Code', 'Female', 'M_M1', 'M_MX', 'N1', 'N2', 'N3', 'NX', 'S1', 'S2A', 'S2B', 'S3A', 'S3B', 'S3C', 'S3N', 'S0']
response = "DX_Death"

# With y as factor
h2oset["DX_Death"] = h2oset["DX_Death"].asfactor()
```

Split dataset into train & valid

```python
# Split the dataset into a train and valid set:
train, valid = h2oset.split_frame(ratios=[.8], seed=42)
```

```python
print(h2oset["DX_Death"].table())
```

### Initial Distributed Random Forests

```python
h2o_drf = H2ORandomForestEstimator(ntrees=10,
                                    max_depth=5,
                                    min_rows=10,
                                    calibrate_model=True,
                                    calibration_frame=valid,
                                    binomial_double_trees=True,
                                    balance_classes=True,
                                    )
```

```python
h2o_drf.train(x=predictors,
               y=response,
               training_frame=train,
               validation_frame=valid)
```

```python
# Eval performance:
perf = h2o_drf.model_performance()
```

```python
perf.plot()
```

```python
from h2o.grid.grid_search import H2OGridSearch
```

## Using grid Search

Parameters to grid search
- Class Balance : Imbalance in classes of response necessitates parameters for adjusting ratio
- Sample Rate Per Class : Further correction of imbalanced data
- Learn Rate : 
- Max Depth :
- Column Sample Rate :

#### On class balance
for parameter \[firstnum, secondnum], firstnum will set the ratio of class 1 kept and secondnum will be ratio of class 0 kept

```python
hyper_params = {'class_sampling_factors': [[1, 0.5], [1,0.2], [1,0.35], [1,0.6]]}
```

```python
class_grid = H2OGridSearch(model = H2ORandomForestEstimator, hyper_params = hyper_params,
                     search_criteria = {'strategy': "Cartesian"})
```

```python
# train using the grid
class_grid.train(x = predictors, y = response, training_frame = train, validation_frame = valid)
```

```python
# sort the grid models by logloss
sorted_grid = class_grid.get_grid(sort_by='logloss', decreasing=False)
print(sorted_grid)
```

For our gridsearch results keeping **0.2** of original class 0 data gave us the best performance

#### On sample_rate parameter
No noticeable difference in sample rate parameter changes -- between 0.5 & 0.6, 0.5 better

```python
srate_params = {'sample_rate': [0.5, 0.6]}
```

```python
gridsr = H2OGridSearch(model = H2ORandomForestEstimator, hyper_params = srate_params,
                     search_criteria = {'strategy': "Cartesian"})
```

```python
# train using the grid
gridsr.train(x = predictors, y = response, training_frame = train, validation_frame = valid)
```

```python
# sort thH2OGridSearchodels by logloss
sorted_gridsr = gridsr.get_grid(sort_by='logloss', decreasing=False)
print(sorted_gridsr)
```
