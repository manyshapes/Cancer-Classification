# Aquiring SEER\*Data through Database

This is a walkthrough for gaining *breastcancerdata.csv* used in *Classification.ipynb* 

SEER Research Data is comprised of multiple databases accessed through SEER\*Stat software. The software is a Windows .exe application [available for download](http://seer.cancer.gov/seerstat/download) with registration. [Parallels Desktop](https://www.parallels.com/) is one option for running the Windows on a Mac.

Once available, the application will ask for login info which can be obtained through [submitting a request](https://seer.cancer.gov/seertrack/data/request/).

In SEER\*Stat session data aquisitions are made through defining **Data** used, **Selection** of data, aspects of **Table**, and title of **Output**.

For raw data files with non-aggregate observations in rows a **Case Listing** data pull will be done. Clicking the grid button will initiate the session. 

## Data



From the datasets available `Incidence - SEER 18 Regs Research Data + Hurrican Katrina Impacted Louisiana Cases, Nov 2018 Sub (1975-2016 varying)` is used. This dataset is used for Incidence-Based Mortality & all documented patients have died by the study cut off date. 

Out of the incidence-based mortality datasets available it spans the longest period & contains the most records. 

## Selection

Once dataset is chosen unique subsets maybe selected. Here we're looking at breast cancer so we'll specify, within `Site and Morphology` :

`Site recode ICD-0-3-/WHO 2008` `is = to` `Breast`


This will populate the `Selection Statement` viewer with our specification. After pressing OK our selection window will be as viewed:

*Note that we are only looking at malignant cases of breast cancer to determine cancer morbidity. Benign cases or otherwise will not be included.*

## Table

Defining our table selects the unique variables we are interested in viewing. These can be selected as Columns that will comprise our data file.

## Output

Relatively self explanatory, the Output tab creates a Title for identifying our dataset. We will then press the lightning bolt symbol or `Session` >> `Execute` to obtain the data.
