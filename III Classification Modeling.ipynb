```python
h2o.init()

# Import the cars dataset into H2O:
h2oset = h2o.import_file('75edits.csv')
```

### Setting predictors & response:
**Response Variable** DX_Death--Diagnosis Death--is a binary classification, 1 being a diagnosis caused morbidity, 0 being otherwise. This will be need to be set as a `factor` in h2o ai.

**Feature** Columns used include:   
- Binary variables: `Female` (1 being True), `Surgery` (1 being True)
- Continuous: `MO_Survival` (in Months), `DX_Year` (diagnosis year), `AGE_Code` (within 5 year bins), `N_Tumors` (number of),
- Ordinal: Stages `S1` ... `S3N`, Metastasis `M_M1`, Nodes Positive `N1` ... `N3`

```python
# Setting x & y
predictors = ['N_Tumors','MO_Survival','DX_Year', 'AGE_Code', 'Female', 'M_M1', 'M_MX', 'N1', 'N2', 'N3', 'NX', 'S1', 'S2A', 'S2B', 'S3A', 'S3B', 'S3C', 'S3N', 'S0']
response = "DX_Death"

# With y as factor
h2oset["DX_Death"] = h2oset["DX_Death"].asfactor()
```

```python
# Split the dataset into a train and valid set:
training, test = h2oset.split_frame(ratios=[.8], seed=42)
```

```python
# Split the dataset into a train and valid set:
train, valid = training.split_frame(ratios=[.8], seed=42)
```

### Initial Distributed Random Forests
A baseline model is created to view performance without augmentations.

```python
h2o_drf = H2ORandomForestEstimator(ntrees=10,
                                    max_depth=5,
                                    min_rows=10,
                                    calibrate_model=True,
                                    calibration_frame=valid,
                                    binomial_double_trees=True,
                                    balance_classes=True,
                                    )
```

```python
h2o_drf.train(x=predictors,
               y=response,
               training_frame=train,
               validation_frame=valid)
```

```python
# Eval performance:
perf = h2o_drf.model_performance()
```

```python
perf.plot(server=True)
# plot.savefig('roc.png', dpi = 300, bbox_inches = "tight")
```

## Improving Parameters with Grid Search

Parameters to grid search

Random Forest
- Class Balance : adjusting ratio of sampled data in classes
- Max Depth : branches per tree, default is 20, may lead to overfitting
- Sample Rate : row sampling rate, ordinarily 0.63

#### On class balance
First number indicates sampled amount of class 1--minority class & morbidity classification. Second number used for sampling of null class.

Note the minority class is 1/5 - 1/6 the size of the null class. By setting the null class as a fraction similar to the minority an undersampling occurs which levels the playing field of training.

```python
hyper_params = {'class_sampling_factors': [[1, 0.5], [1,0.2], [1,0.35], [1,0.17]],
               }
```

```python
class_grid = H2OGridSearch(model = H2ORandomForestEstimator, hyper_params = hyper_params,
                     search_criteria = {'strategy': "Cartesian"})
```

```python
# train using the grid
class_grid.train(x = predictors, y = response, training_frame = train, validation_frame = valid)
```

```python
# sort the grid models by logloss
sorted_grid = class_grid.get_grid(sort_by='logloss', decreasing=False)
print(sorted_grid)
```

For our gridsearch results keeping **0.2** of original class 0 data gave us the best performance

#### On sample_rate parameter
No noticeable difference in sample rate parameter changes -- between 0.5 & 0.6, 0.5 better

```python
srate_params = {'sample_rate': [0.5, 0.6, 0.7]}
```

```python
gridsr = H2OGridSearch(model = H2ORandomForestEstimator, hyper_params = srate_params,
                     search_criteria = {'strategy': "Cartesian"})
```

```python
# train using the grid
gridsr.train(x = predictors, y = response, training_frame = train, validation_frame = valid)
```

```python
# sort thH2OGridSearchodels by logloss
sorted_gridsr = gridsr.get_grid(sort_by='logloss', decreasing=False)
print(sorted_gridsr)
```

A lower sampling rate is giving us better results.
